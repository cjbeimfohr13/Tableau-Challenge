```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
may_2020 = "Resources/JC-202005-citibike-tripdata.csv"
jun_2020 = "Resources/JC-202006-citibike-tripdata.csv"
jul_2020 = "Resources/JC-202007-citibike-tripdata.csv"
aug_2020 = "Resources/JC-202008-citibike-tripdata.csv"
```

```python
may_2020_df = pd.read_csv(may_2020)
jun_2020_df = pd.read_csv(jun_2020)
jul_2020_df = pd.read_csv(jul_2020)
aug_2020_df = pd.read_csv(aug_2020)
aug_2020_df
```

```python
combined_CitiBike_df = pd.merge(may_2020_df, jun_2020_df, how='outer')
combined_CitiBike_df2 = pd.merge(combined_CitiBike_df, jun_2020_df, how='outer')
combined_CitiBike_df3 = pd.merge(combined_CitiBike_df2, jul_2020_df, how='outer')
combined_CitiBike_df3 = pd.merge(combined_CitiBike_df3, aug_2020_df, how='outer')

combined_CitiBike_df3.dtypes
```
